#GERAR NOVAS VARIAVEIS
---

Para criar uma nova feature 'Price1' que é o preço de fechamento de amanhâ, você pode fazer isso.

O Shift(-1) movemos uma coluna inteira do preço de fechamento uma linha para cima.

```python
fb[['Open','Close']]
```

```python
fb['Price1'] = fb['Close'].shift(-1)
```

```python
fb.head(5)
```

```python
fb['PriceDiff'] = fb['PriceDiff'] = fb['Price1'] - fb['Close']
fb.head()
```

```python
#podemos calcular o retorno diário
fb['Return'] = fb['PriceDiff'] / fb['Close']
fb.head()
```

#LIST COMPREHENSION
---
Nãõ podemos usar a transformação de colunas para obter a "direção" da diferença entre amanhã e hoje no fechamento das ações, como fizemos anteriormente.  então devo usar a list comprehension.


```python
fb['Direction'] = [1 if fb.loc[ei, 'PriceDiff'] > 0 else -1 for ei in fb.index]
fb.head()
```

#MOVING AVERAGE
---
A média movel é um sinal muito usado na negociação de ações, que carrega o preço médio em um número definido de dias.
Flutuações aleatórias são muito comuns no preço das ações. O preço médio em um período pode suavizar o ruído. Procura-se que seja bom para encontrar a tendência.

```python
fb["Average3"] = (fb['Close'] + fb['Close'].shift(1) +
                  fb['Close'].shift(2)) / 3
```

```python
fb.head()
```

Pandas tem um método de construção para calcular a média móvel em qualquer numero de dias.

- `.rolling()`

Rolling é um metodo muito bom que nos fornece o calculo da janela de rolagem. Podemos usar essa janela de rolagem para calcula a soma média e uniforme.

```python
fb['MA40'] = fb['Close'].rolling(40).mean()
fb['MA200'] = fb['Close'].rolling(200).mean()
fb.head()
```

#Visualização das novas variáveis
---
Primeiro plotamos o preço de fechamento e, em seguida, a média móvel 40 e depois a MA200.

Vocês podem ver que essas médias móveis suavizam o preço de fechamento original. A média móvel 40, em laranja, é mais associada ao preço de fechamento do que a média móvel 200. Assim, chamamos a média móvel 40 de sinal rápido e a MA200 de sinal lento, que reflete o preço em um longo histórico.

Se a MA40 estiver acima da MA200 (MA40 > MA200), alguns traders, que chamamos de trader que seguem tendências, acreditam que o preço da ação subirá por algum tempo.  

```python
fb['Close'].plot()
fb['MA40'].plot()
fb['MA200'].plot()
```

#ESTRATÉGIA DE NEGOCIAÇÃO
---
A média movel de curto período está mais associada à mudança recente do preço das ações, o que chamamos de sinal rápido.

```python
ms['MA10'] = ms['Close'].rolling(10).mean()
```

```python
 ms.head()
```

```python
ms['MA50'] = ms['Close'].rolling(50).mean()
ms.head()
```

```python
ms['Close'].plot(legend=True)
ms['MA10'].plot(legend=True)
ms['MA50'].plot(legend=True)
```

Traçamos o preço de fechamento, MA10 E MA50.Se MA10 >(for maior que) MA50, alguns traders acreditam que o preço da ação subirá nos proximos dias. Caso contrário, o Preço cairá.
Nossa estratégia é que, se a MA10 for maior que MA50,
compraremos e manteremos uma ação. Como alternativa, compraremos uma ação.

Podemos criar uma variável para indicar se compramos uma ação ou não.



```python
ms['Shares'] = [1 if ms.loc[ei, 'MA10'] > ms.loc[ei, 'MA50'] else 0 for ei in ms.index]
ms.head()
```

#lucro diário
---


```python
ms['Close1'] = ms['Close'].shift(-1)
ms.iloc[500:505,:]
```

```python
ms['Profit'] = [ms.loc[ei, 'Close1'] - ms.loc[ei, 'Close'] if ms.loc[ei, 'Shares'] == 1 else 0 for ei in ms.index]
ms.head()
```

Podemos plotar o lucro e descobrir que em alguns dias ganhamos dinheiro. Acima de 0, ganhou dinheiro e abaixo de 0, perdeu dinheiro.

```python
ms['Profit'].plot()
```

#Podemos calcular a riqueza cumulativa
---
Podemos calcular a riquza cumulativa.
Usamos o metodo cumsum( ) para calcular a soma cumulativa e criar uma nova variável Wealthy.

```python
ms['wealth'] = ms['Profit'].cumsum()
ms.head()
```

```python
ms.tail()
```

Verificamos  a parte da cauda do dataframe e verificamos se ganhamos ou perdemos dinheiro.
NO geral, essa estratégia gera cerca de 24 dolares.

```python
print('O montante que você ganhou foi ', ms.loc[ms.index[-2], 'wealth'])
print('O montante que você gastou foi ', ms.loc[ms.index[0], 'Close'])
ms['wealth'].plot()
```

Riqueza cumulativa é o Patrimônio, que na verdade é uma soma de lucro. Para obter esse lucro, precisamos comprar uma ação inicialmente. Portanto, nosso investimento inicial é o preço da ação no primeiro dia. Essa estratégia parece ser muito promissora. Mas há muitas duvidas se voce quiser implementa-la no mercado real.
