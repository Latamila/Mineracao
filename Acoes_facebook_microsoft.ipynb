#GERAR NOVAS VARIAVEIS
---

Para criar uma nova feature 'Price1' que  o preo de fechamento de amanh, voc pode fazer isso.

O Shift(-1) movemos uma coluna inteira do preo de fechamento uma linha para cima.

```python
fb[['Open','Close']]
```

```python
fb['Price1'] = fb['Close'].shift(-1)
```

```python
fb.head(5)
```

```python
fb['PriceDiff'] = fb['PriceDiff'] = fb['Price1'] - fb['Close']
fb.head()
```

```python
#podemos calcular o retorno dirio
fb['Return'] = fb['PriceDiff'] / fb['Close']
fb.head()
```

#LIST COMPREHENSION
---
N podemos usar a transformao de colunas para obter a "direo" da diferena entre amanh e hoje no fechamento das aes, como fizemos anteriormente.  ento devo usar a list comprehension.


```python
fb['Direction'] = [1 if fb.loc[ei, 'PriceDiff'] > 0 else -1 for ei in fb.index]
fb.head()
```

#MOVING AVERAGE
---
A mdia movel  um sinal muito usado na negociao de aes, que carrega o preo mdio em um nmero definido de dias.
Flutuaes aleatrias so muito comuns no preo das aes. O preo mdio em um perodo pode suavizar o rudo. Procura-se que seja bom para encontrar a tendncia.

```python
fb["Average3"] = (fb['Close'] + fb['Close'].shift(1) +
                  fb['Close'].shift(2)) / 3
```

```python
fb.head()
```

Pandas tem um mtodo de construo para calcular a mdia mvel em qualquer numero de dias.

- `.rolling()`

Rolling  um metodo muito bom que nos fornece o calculo da janela de rolagem. Podemos usar essa janela de rolagem para calcula a soma mdia e uniforme.

```python
fb['MA40'] = fb['Close'].rolling(40).mean()
fb['MA200'] = fb['Close'].rolling(200).mean()
fb.head()
```

#Visualizao das novas variveis
---
Primeiro plotamos o preo de fechamento e, em seguida, a mdia mvel 40 e depois a MA200.

Vocs podem ver que essas mdias mveis suavizam o preo de fechamento original. A mdia mvel 40, em laranja,  mais associada ao preo de fechamento do que a mdia mvel 200. Assim, chamamos a mdia mvel 40 de sinal rpido e a MA200 de sinal lento, que reflete o preo em um longo histrico.

Se a MA40 estiver acima da MA200 (MA40 > MA200), alguns traders, que chamamos de trader que seguem tendncias, acreditam que o preo da ao subir por algum tempo.  

```python
fb['Close'].plot()
fb['MA40'].plot()
fb['MA200'].plot()
```

#ESTRATGIA DE NEGOCIAO
---
A mdia movel de curto perodo est mais associada  mudana recente do preo das aes, o que chamamos de sinal rpido.

```python
ms['MA10'] = ms['Close'].rolling(10).mean()
```

```python
 ms.head()
```

```python
ms['MA50'] = ms['Close'].rolling(50).mean()
ms.head()
```

```python
ms['Close'].plot(legend=True)
ms['MA10'].plot(legend=True)
ms['MA50'].plot(legend=True)
```

Traamos o preo de fechamento, MA10 E MA50.Se MA10 >(for maior que) MA50, alguns traders acreditam que o preo da ao subir nos proximos dias. Caso contrrio, o Preo cair.
Nossa estratgia  que, se a MA10 for maior que MA50,
compraremos e manteremos uma ao. Como alternativa, compraremos uma ao.

Podemos criar uma varivel para indicar se compramos uma ao ou no.



```python
ms['Shares'] = [1 if ms.loc[ei, 'MA10'] > ms.loc[ei, 'MA50'] else 0 for ei in ms.index]
ms.head()
```

#lucro dirio
---


```python
ms['Close1'] = ms['Close'].shift(-1)
ms.iloc[500:505,:]
```

```python
ms['Profit'] = [ms.loc[ei, 'Close1'] - ms.loc[ei, 'Close'] if ms.loc[ei, 'Shares'] == 1 else 0 for ei in ms.index]
ms.head()
```

Podemos plotar o lucro e descobrir que em alguns dias ganhamos dinheiro. Acima de 0, ganhou dinheiro e abaixo de 0, perdeu dinheiro.

```python
ms['Profit'].plot()
```

#Podemos calcular a riqueza cumulativa
---
Podemos calcular a riquza cumulativa.
Usamos o metodo cumsum( ) para calcular a soma cumulativa e criar uma nova varivel Wealthy.

```python
ms['wealth'] = ms['Profit'].cumsum()
ms.head()
```

```python
ms.tail()
```

Verificamos  a parte da cauda do dataframe e verificamos se ganhamos ou perdemos dinheiro.
NO geral, essa estratgia gera cerca de 24 dolares.

```python
print('O montante que voc ganhou foi ', ms.loc[ms.index[-2], 'wealth'])
print('O montante que voc gastou foi ', ms.loc[ms.index[0], 'Close'])
ms['wealth'].plot()
```

Riqueza cumulativa  o Patrimnio, que na verdade  uma soma de lucro. Para obter esse lucro, precisamos comprar uma ao inicialmente. Portanto, nosso investimento inicial  o preo da ao no primeiro dia. Essa estratgia parece ser muito promissora. Mas h muitas duvidas se voce quiser implementa-la no mercado real.
